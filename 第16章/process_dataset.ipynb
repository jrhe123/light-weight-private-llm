```python
import json

# ================= 配置 =================
INPUT_FILE = "math_dataset_with_answers.jsonl"
OUTPUT_FILE = "math_dataset_with_answers_new.jsonl"

# 您需要的核心字段
KEYS_TO_KEEP = ["instruction", "input", "thinking", "output"]

# ================= 执行 =================
print(f"开始提取字段: {KEYS_TO_KEEP}")
count = 0

with open(INPUT_FILE, "r", encoding="utf-8") as f_in, \
     open(OUTPUT_FILE, "w", encoding="utf-8") as f_out:
    
    for line in f_in:
        line = line.strip()
        if not line: continue
        
        try:
            # 1. 读取原始数据
            data = json.loads(line)
            
            # 2. 只提取指定字段，原样保留
            # 使用 .get(k, "") 确保如果字段缺失则补空字符串，不报错
            new_entry = {key: data.get(key, "") for key in KEYS_TO_KEEP}
            
            # 3. 写入新文件
            f_out.write(json.dumps(new_entry, ensure_ascii=False) + "\n")
            count += 1
            
        except json.JSONDecodeError:
            print(f"跳过错误行: {line[:50]}...")

print(f"处理完成！")
print(f"共 {count} 条纯净数据已保存至: {OUTPUT_FILE}")
print(f"新文件仅包含字段: {KEYS_TO_KEEP}")
```
