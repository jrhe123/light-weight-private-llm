```python
with open("clean_data.txt", "r", encoding="utf-8") as f:
    raw_data= f.read()

print(raw_data[:100])
```

```python
w_list = list(raw_data)
print(w_list[0], w_list[1])
```

```python
#去掉空格、空字符或回车换行符
result = [i.strip() for i in w_list if i.strip()]
print(len(result))
```

```python
#建立词汇表
#将result中重复的词去掉，然后对剩下的词进行排序
all_words = sorted(set(result))
#print(len(all_words))
vocab_size = len(all_words)
vocab = {token: idx for idx, token in enumerate(all_words)}

for i, (key, val) in enumerate(vocab.items()):
    print({key: val})
    if i>100:
        break
```
